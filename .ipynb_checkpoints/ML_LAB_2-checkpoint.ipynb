```python
import numpy as np
import random
import matplotlib.pyplot as plt

np.random.seed(10) #Stops from changing random values on every run of program

X = np.sort(np.random.uniform(low=0.0,high=1.0,size=500))
Y = np.sin(2*np.pi*X)
noise = np.random.normal(scale=0.3,size=Y.shape)
Fin_op = Y + noise

plt.plot(X,Fin_op)
plt.show()


random.seed(10)
ind = random.sample(range(500),500)
ind_train  = ind[:200]
x_subtrain = X[ind_train]
y_subtrain = Y[ind_train]
#plt.plot(x_subtrain,y_subtrain)
#print(x_subtrain)
#print(y_subtrain)

from sklearn.linear_model import LinearRegression
L1 = LinearRegression()
L1.fit(x_subtrain[:,np.newaxis],y_subtrain)
predictions = L1.predict(X[:,np.newaxis])
plt.plot(X,Y,color='green')
plt.plot(X,predictions)
plt.show()

```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=10)
x_subpoly = poly.fit_transform(x_subtrain[:,np.newaxis])

#print(x_subpoly)
x_poly = poly.fit_transform(X[:,np.newaxis])
L1.fit(x_subpoly,y_subtrain)
pred1 = L1.predict(x_poly)
plt.plot(X,Y,color='green')
plt.plot(X,pred1,color='red')
plt.scatter(x_subtrain,y_subtrain,s=50,facecolors='none',edgecolors='b')
```
